```python
import sys, os
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D

sys.path.append(os.path.abspath('../..'))

from constants import FIG_HEIGHT, LABEL_FONT_SIZE, DPI

# Create figure and axis
fig, ax = plt.subplots(figsize=(FIG_HEIGHT, FIG_HEIGHT))
ax.set_xticks([])
ax.set_yticks([])

# Set limits for the plot
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

# Draw axes
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)

# Add a dotted unit circle
theta = np.linspace(0, 2 * np.pi, 300)
x_circle = np.cos(theta)
y_circle = np.sin(theta)
ax.plot(x_circle, y_circle, linestyle=':', color='black', linewidth=1, label="$\tau$")

# Add an 'X' at the origin and label
ax.text(0, 0, 'X', ha='center', va='center', color='red', label="Start Position", fontsize=2*LABEL_FONT_SIZE)
ax.text(0.1, -0.2, 'Start Position', color='black', fontsize=LABEL_FONT_SIZE)

# Define flag positions and colors
flags_train = [
    ((1, 0), 'goal location task 1', 'red'),
    ((0, 1), 'goal location task 2', 'orange'),
]

flags_test = [
    ((-np.sqrt(2)/2, -np.sqrt(2)/2), 'goal location task 3', 'blue'),
]

# Function to draw a flag
def draw_flag(x, y, color):
    ax.plot([x, x], [y, y + 0.3], color='black', linewidth=3)  # pole
    ax.plot([x, x + 0.2, x], [y + 0.3, y + 0.15, y], color=color, linewidth=3)  # flag

# Draw all flags
for (x, y), _, color in flags_train:
    draw_flag(x, y, color)

for (x, y), _, color in flags_test:
    draw_flag(x, y, color)

# Create custom legend entries
# Create custom legend entries (including unit circle)
legend_elements = [
    Line2D([0], [0], linestyle=':', color='gray', label='$\mathcal{T}$'),  # unit circle
    Line2D([0], [0], color='red', marker='>', linestyle='None', label='Goal Train Task 0'),
    Line2D([0], [0], color='orange', marker='>', linestyle='None', label='Goal Train Task 1'),
    Line2D([0], [0], color='blue', marker='>', linestyle='None', label='Goal Test Task'),
]



ax.legend(
    handles=legend_elements,
    ncols=2,
    fontsize=LABEL_FONT_SIZE,
    loc='upper center',
    bbox_to_anchor=(0.5, 0)  # Move below the plot
)

# Set aspect ratio and grid
ax.set_aspect('equal')
ax.grid(True, linestyle='--', alpha=0.5)
ax.axis('off')

# Hide top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
fig.savefig("circle_navigation.png", bbox_inches='tight', pad_inches=0.05, dpi=DPI)
```
