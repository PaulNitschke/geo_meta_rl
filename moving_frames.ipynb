## Plotting Utils

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_vector_fields():
    fig, axes = plt.subplots(1, 3, figsize=(24, 8), sharey=True)

    parameters = [(2, 1), (1, 1), (1, 2)]

    for i, (LAMBDA, RHO) in enumerate(parameters):
        radii = np.linspace(0.5, 2, 5)
        VECTOR_LENGTH = 2
        theta = np.linspace(0, 2 * np.pi, 10)
        x = np.concatenate([r * np.cos(theta) for r in radii])
        y = np.concatenate([r * np.sin(theta) for r in radii])

        u = LAMBDA * VECTOR_LENGTH * y
        v = -RHO * VECTOR_LENGTH * x

        ax = axes[i]
        ax.quiver(x, y, u, v)
        for r in radii:
            circle = plt.Circle((0, 0), r, color='gray', linestyle='dotted', fill=False)
            ax.add_artist(circle)
        ax.set_xlabel('$s_1$', fontsize=30)
        if i == 0:
            ax.set_ylabel('$s_2$', fontsize=30)

        ax.set_title(rf'Kernel Dist. $D^R, \lambda={LAMBDA}, \rho={RHO}$', fontsize=30)
        ax.grid(True)
        ax.tick_params(axis='both', which='major', labelsize=25)

    plt.tight_layout()
    plt.savefig('vector_fields.png', bbox_inches='tight', dpi=100)
    plt.show()

# Generate and display the vector fields
plot_vector_fields()




```
