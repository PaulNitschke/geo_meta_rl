## Plotting Utils

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_vector_field(LAMBDA, RHO, show_yticks=True):
    radii = np.linspace(0.5, 2, 5)
    VECTOR_LENGTH = 2
    theta = np.linspace(0, 2 * np.pi, 10)
    x = np.concatenate([r * np.cos(theta) for r in radii])
    y = np.concatenate([r * np.sin(theta) for r in radii])

    u = LAMBDA * VECTOR_LENGTH * y
    v = -RHO * VECTOR_LENGTH * x

    plt.figure(figsize=(8, 8))
    plt.quiver(x, y, u, v)
    for r in radii:
        circle = plt.Circle((0, 0), r, color='gray', linestyle='dotted', fill=False)
        plt.gca().add_artist(circle)
    plt.xlabel('$s_1$', fontsize=30)
    if not show_yticks:
        plt.yticks([])
        plt.ylabel('$s_2$', fontsize=0)
    else:
        plt.ylabel('$s_2$', fontsize=30)
    plt.title(rf'Kernel Dist. $D^R, \lambda={LAMBDA}, \rho={RHO}$', fontsize=30)
    plt.grid(True)
    plt.show()

# Plot the vector fields with different parameters
plot_vector_field(LAMBDA=2, RHO=1)
plot_vector_field(LAMBDA=1, RHO=1, show_yticks=False)
plot_vector_field(LAMBDA=1, RHO=2, show_yticks=False)
```
